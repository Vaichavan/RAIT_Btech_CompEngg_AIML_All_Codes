```python
import os
import numpy as np
import pandas as pd
```

```python
BASE_PATH = 'datasets'
```

```python
sites = os.listdir(BASE_PATH)
print('Total number of sites:', len(sites))
```

```python
df = None
for site in sites:
    if df is not None:
        df = pd.concat((df,pd.read_csv(os.path.join(BASE_PATH, site), skiprows=10)))
    else:
        df = pd.read_csv(os.path.join(BASE_PATH, site), skiprows=10)
df
```

```python
#Making target col as the last col
y=df.pop('GHI (Wh/m2)')
df['GHI (Wh/m2)']=y
```

```python
df.info()
```

```python
#drop wind speed at 3m std dev uncertanity (m/s) as all values are missing
df.drop(columns='Wind Speed at 3m (std dev) Uncertainty (m/s)',inplace=True)
df
```

```python
#Drop any exam with missing target value
df.dropna(how='all', axis='index',subset=['GHI (Wh/m2)'],inplace =True)
df
```

```python
#checking missiong values
for col in df.columns:
    if sum(pd.isnull(df[col]).values)>0:
        print('{0:s} has {1:d} missing values.' .format(col, sum(pd.isnull(df[col]).values)))
```

## option 1:remove all records with any column value missing

```python
# drop records with missing values
df_dropped = df.dropna(how='any' , axis ='index')
df_dropped
```

## option 2 :replace missing values

```python
for col in df.columns:
    if sum(pd.isnull(df[col]).values) >0:
        #Calculate mean of the column
        value = df[col].mean()
        #Replace missing value with calculated value 
        df[col].fillna(value = value, inplace=True)
```

## Calculated value can be mean, median, mode, etc

```python
# Checking missing values
for col in df_dropped.columns:
    if sum(pd.isnull(df_dropped[col]).values) > 0:
        print('{0:s} has {1:d} missing values.'.format(col, sum(pd.isnull(df_dropped[col]).values)))
```

```python
def get_class(ghi):
    if ghi>0.4:
        return 'running'
    elif ghi>=-0.4:
        return 'Monitoring'
    elif ghi>=-1:
        return 'Inspecting'
```

## Generate classes

3 classes:
**Running :** GHI> +0.4,
**Moniotoring :** 0.4<= GHI >=-0.4
**Inspecting:** -0.4< GHI >= -0.1

```python
scaled_df['Class']=scaled_df['GHI (Wh/m2)'].apply(get_class)
scaled_df
```
