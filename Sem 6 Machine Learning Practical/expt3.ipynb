```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Number of data points
n = 100

# Generate X values with a non-zero mean and standard deviation
X = np.random.normal(loc=5, scale=2, size=n)

# Generate Y values using a linear equation with random noise
Y = 2 * X + 3 + np.random.normal(scale=1, size=n)
```

```python
mean_X = np.mean(X)
mean_Y = np.mean(Y)
```

```python
# Calculate variance of X
var_X = np.var(X, ddof=1)  # ddof=1 for sample variance

# Calculate covariance of X and Y
cov_XY = np.cov(X, Y, ddof=1)[0, 1]  # [0, 1] to select the covariance

# Calculate beta (slope)
beta = cov_XY / var_X

# Calculate alpha (intercept)
alpha = mean_Y - beta * mean_X
```

```python
Y_pred = alpha + beta * X
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y, label='Actual Data')  # Scatter plot of actual data
plt.plot(X, Y_pred, color='red', label='Regression Line')  # Plot regression line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# ... (Your data generation code from the previous response) ...

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1, 1), Y, test_size=0.2)  # Reshape X for sklearn

# Initialize Linear Regression model
lr = LinearRegression(fit_intercept=True)

# Train the model
lr.fit(X_train, y_train)

# Make predictions
y_pred = lr.predict(X_test)

# Display coefficients
print("Intercept: {0:2.4f}".format(lr.intercept_))
for ii, coef in enumerate(lr.coef_):
    print("Coeff-{0:2d}: {1:2.4f}".format(ii, coef))

plt.bar(range(len(lr.coef_)), lr.coef_)
plt.xticks(range(len(lr.coef_)))
plt.ylabel('coefficient')
plt.show()

# Calculate and print Mean Squared Error
print("\nMean squared error: {0:2.4f}".format(mean_squared_error(y_test, y_pred)))

# Plot outputs
plt.figure()
plt.scatter(X_test[:, 0], y_test, color="black")  # Use X_test[:, 0] for the x-axis
plt.plot(np.sort(X_test[:, 0]), y_pred[np.argsort(X_test[:, 0])], color="blue", linewidth=3)  # Use y_pred and X_test[:, 0]
plt.show()
```
