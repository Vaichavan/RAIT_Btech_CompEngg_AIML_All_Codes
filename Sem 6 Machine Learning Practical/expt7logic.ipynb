```python
# Import required modules from AIMA-Python
from aima3.logic import FolKB, expr, fol_fc_ask

def main():
    # Create an array to hold clauses
    clauses = []

    # Add first-order logic clauses (rules and facts)
    clauses.append(expr("(American(x) & Weapon(y) & Sells(x, y, z) & Hostile(z)) ==> Criminal(x)"))
    clauses.append(expr("Enemy(Nono, America)"))
    clauses.append(expr("Owns(Nono, M1)"))
    clauses.append(expr("Missile(M1)"))
    clauses.append(expr("(Missile(x) & Owns(Nono, x)) ==> Sells(West, x, Nono)"))
    clauses.append(expr("American(West)"))
    clauses.append(expr("Missile(x) ==> Weapon(x)"))

    # Create a first-order logic knowledge base (KB) with clauses
    KB = FolKB(clauses)

    # Add additional rules and facts with `tell()`
    KB.tell(expr('Enemy(Coco, America)'))
    KB.tell(expr('Enemy(Jojo, America)'))
    KB.tell(expr("Enemy(x, America) ==> Hostile(x)"))

    # Perform forward-chaining inference to find results
    hostile = fol_fc_ask(KB, expr('Hostile(x)'))
    criminal = fol_fc_ask(KB, expr('Criminal(x)'))

    # Print results
    print("Hostile Entities:", list(hostile))
    print("Criminal Entities:", list(criminal))

# Execute main function
if __name__ == "__main__":
    main()
```

```python
from aima3.logic import FolKB, expr, fol_fc_ask

def main():
    # Create an array to hold clauses
    clauses = [
        expr("(American(x) & Weapon(y) & Sells(x, y, z) & Hostile(z)) ==> Criminal(x)"),
        expr("Enemy(Nono, America)"),
        expr("Owns(Nono, M1)"),
        expr("Missile(M1)"),
        expr("(Missile(x) & Owns(Nono, x)) ==> Sells(West, x, Nono)"),
        expr("American(West)"),
        expr("Missile(x) ==> Weapon(x)")
    ]

    # Create a first-order logic knowledge base (KB) with clauses
    KB = FolKB(clauses)

    # Add additional rules and facts with `tell()`
    KB.tell(expr("Enemy(Coco, America)"))
    KB.tell(expr("Enemy(Jojo, America)"))
    KB.tell(expr("Enemy(x, America) ==> Hostile(x)"))

    # Perform forward-chaining inference
    hostile = list(fol_fc_ask(KB, expr("Hostile(x)")))
    criminal = list(fol_fc_ask(KB, expr("Criminal(x)")))

    # Print YES or NO based on the results
    print("Hostile?", "YES" if hostile else "NO")
    print("Criminal?", "YES" if criminal else "NO")

# Execute main function
if __name__ == "__main__":
    main()
```
