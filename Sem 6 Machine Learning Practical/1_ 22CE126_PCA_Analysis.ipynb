## **PCA Analysis22CE1261**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 1. Standardize the dataset
# Assuming your data is in a Pandas DataFrame called 'df'
# Select only numeric columns for standardization
numerical_cols = df.select_dtypes(include=['number']).columns
x = df[numerical_cols].values  # Convert selected columns to NumPy array

x = StandardScaler().fit_transform(x)  # Standardize the data

# 2. Import PCA and choose the number of components
pca = PCA(n_components=3)  # Choose 3 principal components

# 3. Apply PCA to the standardized data
principalComponents = pca.fit_transform(x)
```

```python
# Print the shape of the principal components array
print("Shape of principal components:", principalComponents.shape)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the principal components
ax.scatter(principalComponents[:, 0], principalComponents[:, 1], principalComponents[:, 2])

# Set labels for the axes
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

# Set the title of the plot
ax.set_title('PCA Visualization')

# Display the plot
plt.show()
```

```python
# prompt: 2d plot

import matplotlib.pyplot as plt

# Assuming 'principalComponents' is your array of principal components
plt.scatter(principalComponents[:, 0], principalComponents[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D PCA Visualization')
plt.show()
```

```python
# Calculate the variance ratio explained by each principal component
explained_variance_ratio = pca.explained_variance_ratio_

# Print the variance ratio
print("Explained Variance Ratio:", explained_variance_ratio)
```
