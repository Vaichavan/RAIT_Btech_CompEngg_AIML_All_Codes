```python
import pandas as pd
import numpy as np

# Sample data sets
data_set_1 = [10, 12, 23, 23, 16, 23, 21, 16]
data_set_2 = [32, 45, 12, 67, 89, 21, 34, 90]

# Convert to pandas DataFrame
df = pd.DataFrame({
    'DataSet1': data_set_1,
    'DataSet2': data_set_2
})

# Compute Standard Deviations
std1 = np.std(data_set_1, ddof=1)  # sample std deviation
std2 = np.std(data_set_2, ddof=1)

print("📊 Standard Deviation of DataSet1:", round(std1, 2))
print("📊 Standard Deviation of DataSet2:", round(std2, 2))

# Justification
if std1 > std2:
    print("\n🔍 DataSet1 has a larger spread.")
elif std1 < std2:
    print("\n🔍 DataSet2 has a larger spread.")
else:
    print("\n🔍 Both data sets have equal spread.")

# Compute Covariance
cov_matrix = np.cov(data_set_1, data_set_2)
cov_value = cov_matrix[0, 1]

print("\n📈 Covariance between DataSet1 and DataSet2:", round(cov_value, 2))
```
